```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import matplotlib
# So it does not open them in other windows, % is ipthon syntax
%matplotlib inline 
column_names = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model_year', 'origin', 'car_name']
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data'
car_data = pd.read_csv(url, sep='\s+', header=None, na_values='?')
car_data.columns = column_names
# ; in ipthon squeltches the data
pd.scatter_matrix(car_data, figsize=(16, 16));
```
